```python
def makeMel(precalculated_files,file):
    if os.path.isfile(file):
        name = Path(file).stem
        if(name+'.png' in precalculated_files):
            # print("Skipped %d\r" % name)
            # counter+=1
            return False
        try:
            y, sr = librosa.load(file)
            melspectrogram = librosa.feature.melspectrogram(y=y, sr=sr,  n_mels=128, fmax=8000)
            # save_melspectrogram(melspectrogram, precalculated_dir+mode+'/'+name)
            librosa.display.specshow(librosa.power_to_db(melspectrogram, ref=np.max), y_axis='mel', fmax=8000, x_axis='time')
            plt.axis('off')
            plt.savefig(precalculated_dir+mode+'/'+name, bbox_inches = 'tight', pad_inches = 0, vmin=0, vmax=255)
            # plt.clf()
            return True

        except Exception as e:
            print("failed...",e)
            # counter += 1
            return False
```

```python

precalculated_files = os.listdir(precalculated_dir+mode)
ts = time.time()
counter = 0
makemel_par = partial(makeMel, precalculated_files)

for file in glob.iglob("fma_small/fma_small/**", recursive=True):
    status = makemel_par(file)
    counter += 1
    if status:
        print("Processed %d" % counter)
    else:
        print("Skipped %d" % counter)
        
## DO MULTIPROCESSING RIGHT    
        
# with ThreadPoolExecutor() as executor:
#     for status in executor.map(makemel_par, glob.iglob("fma_small/fma_small/**", recursive=True)):
#         counter += 1
#         if status:
#             print("Processed %d" % counter)
#         else:
#             print("Skipped %d" % counter)

logging.info('Took %s', time.time() - ts)
```
