```python
import numpy as np
import matplotlib.pyplot as plt
import glob
from pathlib import Path
import librosa.display
from concurrent.futures import ProcessPoolExecutor
```

```python
precalculated_files = "precalculated_files_mel/fma_small/"
```

```python
for file in glob.iglob("precalculated_files_mel/fma_numpy/*"):
    name = Path(file).stem
    if(name+'.png' in precalculated_files):
        print("Exists",name)
    melspectrogram = np.load("precalculated_files_mel\\fma_numpy\\%s.npy" % name)
    librosa.display.specshow(librosa.power_to_db(melspectrogram, ref=np.max), x_axis='s', y_axis='mel', cmap='gray_r')
    plt.axis('off')
    plt.savefig(precalculated_files+name,bbox_inches = 'tight', pad_inches = 0, vmin=0, vmax=255)
    print("Saved",name )
```

```python
with ProcessPoolExecutor() as executor:
    for status in executor.map(savemel, glob.iglob("precalculated_files_mel/fma_numpy/*")):
        counter += 1
        if status:
            print("Processed %d" % counter)
        else:
            print("Skipped %d" % counter)
```

```python
def savemel(file):
    name = Path(file).stem
    if(name+'.png' in precalculated_files):
        print("Failed",name)
        return False
    melspectrogram = np.load("precalculated_files_mel\\fma_numpy\\%s.npy" % name)
    power_to_db = librosa.power_to_db(melspectrogram, ref=np.max)
    librosa.display.specshow(power_to_db, x_axis='s', y_axis='mel', cmap='gray_r')
    plt.axis('off')
    plt.savefig(precalculated_files+name,bbox_inches = 'tight', pad_inches = 0, vmin=0, vmax=255)
    print("Saved",name )
```
